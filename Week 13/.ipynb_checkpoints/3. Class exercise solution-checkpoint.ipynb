# Class Exercise Solution

```python
class Item:
    def __init__(self, item_id, name, quantity, price_per_unit):
        self.item_id = item_id
        self.name = name
        self.quantity = quantity
        self.price_per_unit = price_per_unit

    def __str__(self):
        return f'ID: {self.item_id}, Name: {self.name}, Quantity: {self.quantity}, Price per unit: {self.price_per_unit}'


class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f'Item {item.name} added to inventory.')

    def update_item(self, item_id, name=None, quantity=None, price_per_unit=None):
        for item in self.items:
            if item.item_id == item_id:
                if name:
                    item.name = name
                if quantity:
                    item.quantity = quantity
                if price_per_unit:
                    item.price_per_unit = price_per_unit
                print(f'Item {item_id} updated.')
                return
        print(f'Item {item_id} not found.')

    def view_items(self):
        if not self.items:
            print("No items in inventory.")
        else:
            for item in self.items:
                print(item)

    def delete_item(self, item_id):
        self.items = [item for item in self.items if item.item_id != item_id]
        print(f'Item {item_id} deleted from inventory.')

    def search_item(self, search_term):
        for item in self.items:
            if item.name == search_term or item.item_id == search_term:
                print(item)
                return
        print(f'Item {search_term} not found.')


def main():
    inventory = Inventory()

    while True:
        print("\nInventory Management System")
        print("1. Add Item")
        print("2. Update Item")
        print("3. View All Items")
        print("4. Delete Item")
        print("5. Search Item")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            item_id = input("Enter item ID: ")
            name = input("Enter item name: ")
            quantity = int(input("Enter item quantity: "))
            price_per_unit = float(input("Enter price per unit: "))
            item = Item(item_id, name, quantity, price_per_unit)
            inventory.add_item(item)

        elif choice == '2':
            item_id = input("Enter item ID to update: ")
            name = input("Enter new name (leave blank to keep current): ")
            quantity = input("Enter new quantity (leave blank to keep current): ")
            price_per_unit = input("Enter new price per unit (leave blank to keep current): ")
            inventory.update_item(
                item_id,
                name if name else None,
                int(quantity) if quantity else None,
                float(price_per_unit) if price_per_unit else None
            )

        elif choice == '3':
            inventory.view_items()

        elif choice == '4':
            item_id = input("Enter item ID to delete: ")
            inventory.delete_item(item_id)

        elif choice == '5':
            search_term = input("Enter item name or ID to search: ")
            inventory.search_item(search_term)

        elif choice == '6':
            print("Exiting Inventory Management System.")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```

**Explanation:** <br>

**Item Class:** This class represents an item in the inventory with attributes for ID, name, quantity, and price per unit. The __str__ method provides a string representation of the item for easy printing. <br>

**Inventory Class:** This class manages the list of items. It includes methods to add, update, view, delete, and search for items. <br>

**Main Function:** This function provides a text-based user interface. It continuously prompts the user to choose an action (add, update, view, delete, search, or exit) and calls the corresponding method on the Inventory object.
