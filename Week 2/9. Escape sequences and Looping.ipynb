Escape sequences are special characters or combinations that are used to represent non-printable or illegal characters in a string. Escape sequences are prefixed by a backlash \ followed by the character(s) you want to insert.
<br>



```python
print("Hello World")
print("Hello\nWorld")
print("Hello\tWorld")
print("This is a backslash: \\")
print("\x48")
```

# Looping through strings
We can loop through strings using for loops, the range function enumerate function. Looping is the process of iterating through each character in the string. With this, you can perform operations on individual characters or process the string as a sequence. 
<br> <br>
**For loop**

```python
text = "Python"
for char in text:
    print(char)
```

**Range function**
<br>
With the range function, you can iterate over the indices of a string and use indexing to access the characters.

```python
text = "Python"
for i in range(len(text)):
    print(text[i])
```

**Enumerate function**
<br>
Used to iterate over both indices and values of a string.

```python
text = "Python"
for index, char in enumerate(text):
    print(f"Index: {index}, Character: {char}")
```
