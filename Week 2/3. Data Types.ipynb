**Setting data types** 
<br>
You set the data type in Python by assigning a value to a variable:

```python
x = 5           # x is an integer
x = "Hello"     # x is a string
x = [1, 2, 3]   # x is a list
```

You can also specify the data type of a variable either when declaring it or when using it. <br>
 
Specifying a variable type during declaration (casting):

```python
x = str(20)                                 # Without specifying the data type, x will be an integer
x = list(('apples', 'oranges', 'grapes'))   # X returns a list
x = int(10.5)                               # Without specifying the data type, x will be a floating point number
```

**Type conversion:** <br>
You can convert from one type to another even after the variable has been declared.

```python
x = 2       #int
y = 3.5     #float
z = 2j      #complex

#from int to float
a = float(x)

#from float to int
b = int(y)

#from int to complex
c = complex(x)

print(a)
print(b)
print(c)
```

You can check a variable’s type by using the built-in function type()

```python
x = 20
print(type(x))

x = str(20)
print(type(x))
```
