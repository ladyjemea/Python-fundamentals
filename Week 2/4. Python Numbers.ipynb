# Python numbers

The three main Python numeric types include: **int, float, complex**. The main number types we will work with in this course are Integers (positive or negative whole numbers) and Floating-Point Numbers (positive or negative numbers with a decimal).

```python
x = 10      # integer
y = -15.5   # floating point number
z = 2j      # complex number
```

**Basic math** <br>

Arithmetic operators are used in Python when numeric values perform mathematical operations. <br> <br>



<br>
Examples:

```python
x = 10
y = 2


#Addition:
result = x + y  #result = 12

#Subtraction:
result = x - y  #result = 8

#Multiplication:
result = x * y  #result = 20

#Division:
result = x / y  #result = 5.0

#Modulus:
result = x % y  #result = 0

#Exponentiation:
result = x ** y  #result = 100
result = pow(x, y)  #result = 100

#Floor division:
result = x // y  #result = 5
```

The modulus operator can be used to check if one number is evenly divisible by another. If the output is 0, then the left operand is evenly divisible by the right operand. We can also check if a number is even or odd. If the right operand is 2 and the result is 0, the left operand is an even number. 

```python
print(7 % 3)   #7 is not evenly divisible by 3. The remainder after division is 1

print(8 % 2)   #8 is evenly divisible by 2 which means it is an even number. The remainder after division is 0
```

Order of operations: <br>

When we have more than one arithmetic operation to perform in a single expression, a specific order of operations needs to be used to ensure correctness referred to as the PEMDAS acronym. <br>

1. Parentheses
2. Exponents (Powers and roots)
3. Multiplication and Division
4. Addition and Subtraction

```python
print(2 + 5 * 8)     #multiplication is performed before addition
print((2 + 5) * 8)   #Using parenthesis forces the addition to be done first
```
