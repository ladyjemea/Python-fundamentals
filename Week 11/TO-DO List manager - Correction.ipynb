# TO-DO LIST MANAGER - Correction

Some corrections/updates I recommend: <br>

- The file is not closed properly. It should be file.close() instead of file.close. <br>
- The code to generate task positions and the dictionary mapping is repeated. This can be refactored into a function. <br>
- Use with statements for file operations to ensure files are properly closed even if an error occurs. <br>
- The process of converting tasks to a list and splitting by newline can be simplified. <br>
- Recursion is used for re-prompting in case of invalid input. This can be changed to a loop for better efficiency and readability. <br>
- Improve error handling for invalid inputs (e.g., non-numeric values where numbers are expected).

```python
#!/usr/bin/env python
# coding: utf-8

# Managing a TO DO LIST THAT IS STORED IN AS A .txt file

def get_file_content(path, file_title):
    with open(f"{path}/{file_title}", 'r') as file:
        return [line.strip() for line in file]

def display_tasks(tasks):
    print("\n Your TO DO list is as follows : \n")
    for i, task in enumerate(tasks, 1):
        print(f"{i}.) {task}")

def my_to_do_list(to_do_list):
    while True:
        response = input("\nAre there any tasks in this list you have already completed? (yes or no): ")
        if response.lower() in ['yes', 'no']:
            break
        else:
            print("\n You can only respond with 'yes' or 'no'")
    
    if response.lower() == 'yes':
        to_do_list = remove_task(to_do_list)
    
    to_do_list = add_task(to_do_list)
    update_task(to_do_list)

def remove_task(to_do_list):
    try:
        num_task_done = int(input("Enter the number of tasks you have already completed: "))
        ranks_to_be_del = []
        for _ in range(num_task_done):
            while True:
                try:
                    rank = int(input("Enter the position of a completed task: "))
                    if 1 <= rank <= len(to_do_list):
                        ranks_to_be_del.append(rank - 1)
                        break
                    else:
                        print("Invalid position, try again.")
                except ValueError:
                    print("Invalid input, please enter a number.")
        
        to_do_list = [task for i, task in enumerate(to_do_list) if i not in ranks_to_be_del]
        return to_do_list
    
    except ValueError:
        print("Error! The value you just entered is not numeric, try again \n")
        return remove_task(to_do_list)

def add_task(to_do_list):
    while True:
        response = input("Do you want to add new tasks to this list? (yes or no): ")
        if response.lower() in ['yes', 'no']:
            break
        else:
            print("You can only respond with 'yes' or 'no'")
    
    if response.lower() == 'yes':
        try:
            num_of_new_tasks = int(input("How many tasks do you want to add to your TO DO LIST?: "))
            for _ in range(num_of_new_tasks):
                new_task = input("Write in a task you want to add to your list: ")
                to_do_list.append(new_task)
        except ValueError:
            print("Invalid input, please enter a number.")
            return add_task(to_do_list)
    
    return to_do_list

def update_task(to_do_list):
    with open(f"{path}/{file_title}", 'w') as file:
        for task in to_do_list:
            file.write(task + '\n')
    
    print("\n Your updated TO DO list is as follows : \n")
    display_tasks(to_do_list)

if __name__ == "__main__":
    path = input("ENTER the Path where the text file containing your ToDoList is found: ")
    file_title = input("Enter the TITLE of this text file with its extension (e.g., example.txt): ")
    
    to_do_list = get_file_content(path, file_title)
    display_tasks(to_do_list)
    my_to_do_list(to_do_list)
```
