# Python Operators

Operators are keywords or symbols used to perform operations on variables and values. Operators allow you to perform calculations and manipulations with different types of data. The different types of operators which Python supports are:
<br> <br>
Arithmetic Operators <br>
Assignment Operators <br>
Comparison Operators <br>
Logical Operators <br>
Membership Operators <br>
Bitwise Operators <br>

**Arithmetic Operators**
<br>
Arithmetic operators are used in Python when numeric values perform mathematical operations. <br> <br>



<br>
Examples:

```python
x = 10
y = 2

#Addition:
result = x + y  #result = 12

#Subtraction:
result = x - y  #result = 8

#Multiplication:
result = x * y  #result = 20

#Division:
result = x / y  #result = 5.0

#Modulus:
result = x % y  #result = 0

#Exponentiation:
result = x ** y  #result = 100
result = pow(x, y)  #result = 100

#Floor division:
result = x // y  #result = 5
```

**Assignment Operators**
<br>
These Operators are used to assign values to variables. They can also be used as a shorthand way of performing an operation and updating the variable in the same step. <br><br>



<br>
Examples:

```python
#Assign
x = 10

#Add and assign
x += 3

#Subtract and assign
x -= 3

#Multiply and assign
x *= 3

#Divide and assign
x /= 3

#Floor divide and assign
x //= 3

#Modulus and assign
x %= 3

#Exponentiate and assign
x **= 3

#Bitwise AND Operator 
a = 5 # Binary: 0101
b = 3 #Binary: 0011

result = a & b #0001 (decimal 1)

#Bitwise OR Operator 
a = 5 # Binary: 0101
b = 3 #Binary: 0011

result = a | b #0111 (decimal 7)

#Bitwise XOR Operator 
a = 5 # Binary: 0101
b = 3 #Binary: 0011

result = a ^ b #0110 (decimal 6)

# Bitwise right shift assignment
x = 16   # Binary: 10000
x >>= 2  # Shift right by 2 positions x = x >> 2


# Bitwise left shift assignment
x = 5    # Binary: 0101
x <<= 2  # Shift left by 2 positions x = x << 2
```

**Comparison Operators**

These are used to compare two values or expressions. The result is a Boolean value (True or False) based on the comparison result. <br><br>



<br>
Examples:

```python
x = 10
y = 2

#Equal
result = (x == y) #False

#Not equal
result = (x != y) #True

#Less than
result = (x < y) #False

#Greater than
result = (x > y) #True

#Less than or equal to
result = (x <= y) #False

#Greater than or equal to
result = (x >= y) #True
```

**Logical Operators**

These are used to perform logical operations on Boolean values. They are used to combine conditional statements and return a Boolean result. Logical operators are used in conditional statements and loops to control the flow of a program based on certain conditions. They help in decision making and code structure. The three main logical operators in Python are: and (logical AND), or (logical OR) and not (logical NOT). <br><br>



<br>
Examples:

```python
#And
x = 10
print(x > 5 and x < 20)

#Or
x = 10
print(x > 5 or x < 9)

#Not
x = 10
print(not(x > 5 and x < 20))
```

**Identity Operators**


Identity operators are used to compare objects, to see if they are the same object with the same memory location, and not if they are equal. They check if two variables refer to the exact same memory location. The two identity operators in Python are:
<br><br>
**Is:**
<br>
Returns True if both variables/operands refer to the same object, otherwise False.
<br>

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)
print(x is z)
print(x == y)
```

**Is not:**
<br>
Returns True if the variables/operands do not refer to the same object, otherwise False.

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is not y)
print(x is not z)
print(x != y)
```

**Membership Operators**


These operators are used to test whether a value is a member of a sequence and return a Boolean value (True or False) based on whether the specified value is present in the sequence. Sequences can be lists, sets or tuples. They are particularly used in conditional statements to check if a particular value exists in a sequence before performing further actions. The two membership operators in Python are:
<br><br>
**In:**
<br>
Returns True if a specified value is found in the sequence, otherwise returns False.

```python
x = [1, 2, 3]
result = 3 in x

print(result)
```

**Not in:**
<br>
Returns True if a specified value is not found in the sequence, otherwise returns False.

```python
x = [1, 2, 3]
result = 3 not in x

print(result)
```

**Bitwise Operators:**


These are operators used to compare binary numbers. They perform operations at the binary level, manipulating individual bits of integers. They are used mostly in low-level programming, data manipulation and optimization. The bitwise operators in Python are:
<br><br>

**Bitwise AND (&):**
<br>
Compares each bit of the first operand to the corresponding bit of the second operand. If both bits are 1, the corresponding result is set to 1, otherwise 0.

```python
a = 5    # 0101
b = 3    # 0011
result = a & b  # 0001 (decimal 1)

print(result)
```

**Bitwise OR (|):**
<br>
Compares each bit of the first operand to the corresponding bit of the second operand. If either bit is 1, the result is set to 1; otherwise, 0.

```python
x = 5    # 0101
y = 3    # 0011
result = x | y  # 0111 (decimal 7)

print(result)
```

**Bitwise XOR (^):**
<br>
Compares each bit of the first operand to the corresponding bit of the second operand. If the bits are different, the corresponding result bit is set to 1. If the bits are the same, the result bit is set to 0.

```python
m = 5    # 0101
n = 3    # 0011
result = m ^ n  # 0110 (decimal 6)

print(result)
```

**Bitwise NOT (~):**
<br>
Inverts all bits if the operand.

```python
p = 5    # 0101
result = ~p  # -(0101 + 1)

print(result)
```

#### Bitwise Left Shift (<<):
<br>
Shifts the bits of the first operand to the left by the number of positions specified by the second operand. New bits on the right are filled with 0.

```python
q = 5    # 0101
result = q << 2  # 10100 (decimal 20)

print(result)
```

#### Bitwise Right Shift (>>):
<br>
Shifts the bits of the first operand to the right by the number of positions specified by the second operand. For signed numbers, the new bits on the left depend on the sign of the number in most languages (filled with 0 for unsigned numbers and with the sign bit for signed numbers).

```python
r = 16   # 10000
result = r >> 2  # 00100 (decimal 4)

print(result)
```

**Operator Precedence:**
<br>
This describes the order in which operations are performed, with the highest precedence at the top.
