## Control flow:
<br>
Control flow in Python refers to the order in which the statements and instructions of a program are executed. It determines the flow of execution based on certain conditions, loops and function calls. Python provides several control flow constructs to enable you to direct the flow of your program. The primary control flow structures include:
<br>
-	Sequential Execution:
<br>
The default mode of execution where statements are executed one after the other in the order they appear in the code.
<br>
-	Conditional statements (if, elif, else):
<br>
Used for decision-making. The execution of code depends on whether a certain condition is true or false.
<br>
-	Loops (for, while):
<br>
Allow you to repeat a block of code multiple times.
<br>
-	Break and Continue Statements:
<br>
Break: Terminates the loop prematurely.
<br>
Continue: Skips the rest of the loop and moves to the next iteration.
<br>
-	Exception handling (try, except, finally):
<br>
Used to handle errors and exceptions that may occur during program execution.

**If statements:**
<br>
If statements are used for decision-making and controlling the flow of a program. They allow you to execute certain blocks of code based on whether a given condition is true or fasle.
<br>
**If statement:**
<br>

<br>
The condition is an expression that evaluates to either True or False. If the condition is True, the indented block of code is executed; otherwise, it is skipped, and the program continues sequentially.
<br>

Example:

```python
x = 10

if x > 0:
    print("x is positive")
    
print('end of program')
```

**Else:**
<br>
You can include an optional else block to specify the code that should be executed if the condition is False.
<br>

<br>

Example:

```python
x = -1

if x > 0:
    print("x is positive")
else:
    print('x is either zero or negative')
    
print('end of program')
```

**Elif:**
<br>
Additionally, you can use elif (short for else if) to check multiple conditions in sequence.



<br>
Example:

Comparison and logical operators are used in the conditions since they return a Boolean which will determine the flow of the program. These conditions help create more complex decision-making structures in your code.
